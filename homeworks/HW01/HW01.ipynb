### Размерность Numpy массивов
(некоторые типы данных уже были рассмотренны при создании массивов)

```python
# создадим Numpy массив в виде первых 100 натуральных чисел и узнаем его размерность
hundred = np.arange(1, 101)

print(f"Созданный массив: \n {hundred} \n")
print(f"Размерность массива: {hundred.shape}")
print(f"Кол-во элементов массива {len(hundred)}")
```

```python
# рассмотрим новые массивы, полученные изменением размерности исходного
print(f"Массив из 5 строк и 20 столбцов: \n {hundred.reshape((5, 20))} \n")
print(f"Размерность: {hundred.reshape((5, 20)).shape}")
```

```python
print(f"3D массив из 5 матриц размера 5x4: \n {hundred.reshape((5, 5, 4))} \n")

print(f"Размерность: {hundred.reshape((5, 5, 4)).shape}")
```

```python
# Рассмотрим массивы с N измерениями, где N-1 измерений заданы,
# а одно определяется автоматически с помощью -1
print(f"Массив размерности {hundred.reshape((-1, 1)).shape}: \n {hundred.reshape((-1, 1))} \n")
```

```python
print(f"Массив размерности {hundred.reshape((10, -1, 10)).shape}: \n {hundred.reshape((10, -1, 10))} \n")
```

```python
print(f"Массив размерности {hundred.reshape((1, -1, 5)).shape}: \n {hundred.reshape((1, -1, 5))} \n")
```

```python
try:
    print(f"Пример использования более одной неизвестной размерности: \
      \n {hundred.reshape((-1, -1, 5))}")
except ValueError as e:
    print(e.args)

try:
    print(f"Пример невозможного изменения размерности: \
      \n {hundred.reshape((5, 5, 5))}")
except ValueError as e:
    print(e.args)
```

### Примеры индексирования и срезов

```python
# Отдельные элементы и срезы будем брать из hundred
# с изменённой размерностью (1, 20, 5)
hundred_cut = hundred.reshape((1, -1, 5))
print(hundred_cut.dtype, "\n", hundred_cut)
```

```python
print(f"Первый элемент последней строки: {hundred_cut[0][-1][0]}")
print(f"Первая строка: {hundred_cut[0, 0]}")
print(f"Второй столбец с конца: {hundred_cut[0, :, -2]}")
print(f"Медиана десятой строки: {hundred_cut[0, 9, hundred_cut.shape[-1]//2]}")
print(f"Трейий и пятый элементы каждой чётной строки: \n {hundred_cut[0, ::2, 2::2]}")
```

```python
try:
    print(f"Пример некорректного индекса: \
      \n {hundred_cut[0, 0.5]}")
except IndexError as e:
    print(e.args)
```

### Примеры векторных операций и базовых статистик

```python
something = hundred_cut[0, ::3, ::2]
something
```

```python
# результат применения тригонометрических функций
print(f"Синус: \n {np.sin(something)}")
print(f"Гиперболический тангенс: \n {np.tanh(something)}")
print(f"Гиперболический арксинус: \n {np.arcsinh(something)}")
print(f"Арккосинус: \n {np.arccosh(something)}")
```

```python
# статистики
print(f"Среднее каждой строки: \n {np.mean(something, axis=1)}")
print(f"Медиана каждой строки: \n {np.median(something, axis=1)}")
print(f"Стандартное отклонение для каждого столбца: \n {np.std(something, axis=0)}")
print(f"Сумма по каждой строке: \n {np.sum(something, axis=1)}")
print(f"Сумма по всему массиву: \n {np.sum(something)}")
print(f"Максимум по каждой строке: \n {np.max(something, axis=1)}")
```

```python
# операции с массивами
print(f"Сложение (поэлементно): \n {something + np.sin(something)}\n")
print(f"Умножение (поэлементно): \n {something * np.pow(something, 2)}\n")
print(f"Объединение по столбцам: \n {np.hstack([something, -0.25*something])}\n")
print(f"Объединение по строкам: \n {np.vstack([np.sin(something), np.pow(something, 2)]).round(4)}\n")
print(f"Умножение (векторное): \n {something @ something.reshape(3, -1)}")
```

```python
try:
    print(f"Пример некорректного векторного умножения (несоместимые размерности): \
      \n {something @ something}")
except ValueError as e:
    print(e.args)
```
